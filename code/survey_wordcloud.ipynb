```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
import matplotlib.pyplot as plt
```

# For Sheet 1

```python
# 读取数据, 需要在同一个文件夹下
dt = pd.read_excel("For Snapshot_WY.xlsx", sheet_name=0)
dt.columns = ['Topic', 'freq']
# dt['freq'] = (dt['freq'] - dt['freq'].min()+1) / (dt['freq'].max() - dt['freq'].min()+1) * 10000 # 将差异放大
dt
```

## color map style
- https://matplotlib.org/1.2.1/mpl_examples/pylab_examples/show_colormaps.pdf
- colormap 决定了颜色风格

```python
topic_freq = dt.set_index('Topic')['freq'].to_dict() # 数据转换

wordcloud = WordCloud(scale=30, width=400, height=200, relative_scaling=1, prefer_horizontal=0.8, mask = None, 
                      min_font_size=1, max_font_size=30, background_color='white', colormap='Purples');
wordcloud.fit_words(topic_freq) # 把数据输入 wordcloud 生成器
plt.figure(figsize=(40,20)); # 图片大小
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off");
```

```python
topic_freq = dt.set_index('Topic')['freq'].to_dict() # 数据转换

wordcloud = WordCloud(scale=30, width=400, height=200, relative_scaling=1, prefer_horizontal=0.8, mask = None, 
                      min_font_size=1, max_font_size=30, background_color='white', colormap='tab20_r');
wordcloud.fit_words(topic_freq) # 把数据输入 wordcloud 生成器
plt.figure(figsize=(40,20)); # 图片大小
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off");
```

# For Sheet 2

```python
# 读取数据, 需要在同一个文件夹下
dt = pd.read_excel("For Snapshot_WY.xlsx", sheet_name=1)
dt.columns = ['Topic', 'freq']
```

```python
text = " ".join(dt.Topic.tolist())

wordcloud = WordCloud(scale=3,  width=1000, height=1000, mask = None,background_color='white', contour_width=1, contour_color='steelblue', stopwords=STOPWORDS,collocations = True,colormap='plasma');
wordcloud.generate(text) # 把文本输入 wordcloud 生成器
plt.figure(figsize=(10,10));
plt.imshow(wordcloud);
plt.axis("off");
```

```python
topic_freq = dt.set_index('Topic')['freq'].to_dict() # 数据转换

wordcloud = WordCloud(scale=3, mask = None, font_step=5, contour_width=1) # 画图参数
wordcloud.generate_from_frequencies(frequencies=topic_freq) # 把数据输入 wordcloud 生成器
plt.figure(figsize=(20,20)); # 图片大小
plt.imshow(wordcloud);
plt.axis("off");
```
