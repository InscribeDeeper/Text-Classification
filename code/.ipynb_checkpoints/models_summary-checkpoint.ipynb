```python
from glob import glob
import pandas as pd
```

```python
output_path = '../Report/v2/'
```

```python
gs = pd.DataFrame()
for file in glob('./GridSearch_result/*.xlsx')[0:]:
    print(file)
    tmp = pd.read_excel(file, index_col=0)
    tmp['file'] = file
    gs = pd.concat([gs, tmp], ignore_index=True, axis=0)
```

```python
gs['rank_of_all'] = gs['mean_test_score'].rank(ascending=False)
gs = gs.sort_values(by='rank_of_all').reset_index()

def clean_filename(x):
    x = x.replace('_cv_result_Classification_svm_nb_baseline','')
    x = x.replace('.\\','')
    x = x.replace('.xlsx','')
    x = x.replace('.ipynb','')
    return x

gs['model_preprocessing'] = gs['file'].apply(lambda x: clean_filename(x))
gs.to_excel(output_path+"gs_all.xlsx")
gs_res = gs.groupby('file').first().sort_values(by='rank_of_all').round(3)
gs_res.columns

```

```python
gs_res.to_excel(output_path+"gs_summary_detail.xlsx")
```

```python
gs_res = gs_res[['model_preprocessing','rank_of_all', 'params', 'mean_fit_time','mean_test_score','std_test_score']]
gs_res.to_excel(output_path+"gs_summary_to_report.xlsx")
```

```python
data = pd.read_csv('bert_clf.csv')
data
```

```python
import matplotlib.pyplot as plt
print("========================================================================")
xx = list(range(1, len(data) + 1))
plt.plot(xx, data.train_acc, color='navy', lw=2, label='Model_train_acc')
plt.plot(xx, data.val_acc, color='darkorange', lw=2, label='Model_val_acc')
plt.title("train_acc v.s. val_acc")
plt.xlabel('Epochs')
plt.ylabel('acc')
plt.title('Finetune BERT training accuracy')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
print("========================================================================")
xx = list(range(1, len(data) + 1))
plt.plot(xx, data.train_loss, color='navy', lw=2, label='Model_train_loss')
plt.plot(xx, data.val_loss, color='darkorange', lw=2, label='Model_val_loss')
plt.title("train_loss v.s. val_loss")
plt.xlabel('Epochs')
plt.ylabel('loss')
plt.title('Finetune BERT training loss')
plt.legend()
plt.show()
```
